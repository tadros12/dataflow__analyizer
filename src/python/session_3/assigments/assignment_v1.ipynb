# Write a Python program to remove an item from a given set

## Example
my_set: {1, 2, 3, 4, 5}

Output: {1, 2, 4, 5}

```python
# Write your answer here

def remove_item(list , item):
    if item in list:    
        list.remove(item)
        print(f"removed {item} from list")
    else:
        print(f"{item} not found in list")
    return list




my_list = [1, 2, 3, 4, 5]
item_to_remove = 3
updated_list = remove_item(my_list, item_to_remove) 
print("Updated list:", updated_list)
```

# Write a Python program to removes multiple items from a given set

## Example
my_set: {1, 2, 3, 4, 5}

items to be removed: {1, 4}

Output: {2, 3, 5}

```python
# Write your answer here


def remove_multiple_items(lst, items):
    for item in items:
        if item in lst:
            lst.remove(item)
            print(f"removed {item} from list")
        else:
            print(f"{item} not found in list")
    return lst


my_list = [1, 2, 3, 4, 5]
items_to_remove = [1,4]
updated_list = remove_multiple_items(my_list, items_to_remove) 
print("Updated list:", updated_list)
```

# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
# Write your answer here

numb = 5

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
result = factorial(numb)
print(f"The factorial of {numb} is {result}")
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here

def sum_even_numbers(lst):
    even_sum = 0
    for num in lst:
        if num % 2 == 0:
            even_sum += num
    return even_sum

numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(f"The sum of even numbers in the list is: {result}")
```

# Create a function that takes a list as input and return reversed list

## Example
lst: [1, 2, 3]

Output: [3, 2, 1]

```python
# Write your answer here

def reverse_list(lst):
    
    reversed_lst = lst[::-1]
    return reversed_lst
my_list = [1, 2, 3]

reversed_list = reverse_list(my_list)
print("Reversed list:", reversed_list)
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
# Write your answer here

def median(lst):
    sorted_lst = sorted(lst)
    n = len(sorted_lst)
    mid = n // 2
    if n % 2 == 0:
        median = (sorted_lst[mid - 1] + sorted_lst[mid]) / 2
    else:
        median = sorted_lst[mid]
    return median

lst = [5,2,7,3,1]
med = median(lst)
print(f"The median of the list is: {med}")
```
