```python
from abc import ABC,abstractmethod
class Animal(ABC):
    
    @abstractmethod
    def move(self):
        pass

class bird(Animal):
    def move(self):
        print("bird")

class cat(Animal):
    def move(self):
        print("cat")
```

```python
a = Animal()
a.move
```
