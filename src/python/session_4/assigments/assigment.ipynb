```python
import os
def thanos_folder(folder_path ,number_of_files):
    
    """ Creates a folder with a specified number of text files and then deletes half of them.
    Args:
        folder_path (str): The path to the folder where files will be created.
        number_of_files (int): The total number of text files to create.
    """
    
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    
    for i in range(number_of_files):
        file_name = f"file_{i+1}.txt"
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'w') as f:
            f.write(f"this is file number {i+1}\n")
        
    half_files = number_of_files // 2
    for i in range(half_files):
        file_name = f"file_{i+1}.txt"
        file_path = os.path.join(folder_path, file_name)
        if os.path.exists(file_path):
            os.remove(file_path)
    
    
    
path = "/run/media/theodoros/E/projects/dataflow__analyizer/folder_toTestRandom$hit"

thanos_folder(path, 10)
```
