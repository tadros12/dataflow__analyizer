```python
encoded_message_split.remove("###")
print(encoded_message_split)
encoded_message_split.remove("6789")
print(encoded_message_split)
```

```python
message_new = encoded_message_split[1]
message_new = message_new.split("@")
meesage_new = message_new[1]
```

```python
final = message_new[1]
```

```python

words = final.split()  

output = []
for word in words:
    reversed_word = word[::-1]   
    output.append(reversed_word)

result = " ".join(output)
print(output)
print(result)
    
```
