```python
import requests
from bs4 import BeautifulSoup
import csv
from time import sleep
from random import uniform
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

base_url = "https://wuzzuf.net/search/jobs"
query = "python"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
number_of_pages = 2

JOB_CARD_CLASS = "css-1gatmva"  # <-- replace with the class you found

with open("wuzzuf_jobs.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Job Title", "Company", "Location", "Posted", "Skills"])

    for page in range(number_of_pages):
        params = {"q": query, "start": page * 15}
        response = requests.get(base_url, headers=headers, params=params, verify=False)
        soup = BeautifulSoup(response.content, "html.parser")

        job_links = soup.select('a.css-o171kl[href*="/jobs/p/"]')

        for link in job_links:
            title = link.get_text(strip=True)
            card = link.find_parent("div", class_=JOB_CARD_CLASS)

            company = location = posted = skills = "N/A"
            if card:
                company_elem = card.find("a", class_=lambda x: x and "css-17s97q8" in x)
                location_elem = card.find("span", class_=lambda x: x and "css-5wys0k" in x)
                posted_elem = card.find("div", class_=lambda x: x and "css-do6t5g" in x)
                skills_elem = card.find("div", class_=lambda x: x and "css-y4udm8" in x)

                company = company_elem.get_text(strip=True) if company_elem else "N/A"
                location = location_elem.get_text(strip=True) if location_elem else "N/A"
                posted = posted_elem.get_text(strip=True) if posted_elem else "N/A"
                skills = skills_elem.get_text(" ", strip=True) if skills_elem else "N/A"

            writer.writerow([title, company, location, posted, skills])

        sleep(uniform(1, 2))
```
