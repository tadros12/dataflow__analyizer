```python
import requests 
from bs4 import BeautifulSoup
import csv


url = "https://books.toscrape.com/"

request = requests.get(url)
soup = BeautifulSoup(request.content, "html.parser")

### scanning for categories ###
categories = []

side_bar = soup.find("div", class_="side_categories")
# print(side_bar)
links = side_bar.find_all("a")
print(len(links))

for a in links[1:]: ## skipping the first link 'Books'
    name = a.text.strip()
    print(name)
    href = a["href"]
    category_url = url + href
    categories.append({"name": name, "url": category_url})


print(categories)





```

```python
for category in categories:
    category_name = category["name"]
    category_url = category["url"]
    
    request = requests.get(category_url)
    soup = BeautifulSoup(request.content, "html.parser")
    
    cards = soup.find_all("article", class_="product_pod")
    
    with open(f"{category_name}.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "Price", "Availability"])
        
        for card in cards:
            title = card.h3.a["title"]
            price = card.find("p", class_="price_color").text
            availability = card.find("p", class_="instock availability").text.strip()
            
            writer.writerow([title, price, availability])
            
            print(f"{title} - {price} - {availability}")
            
            
```
