```python
import requests
from bs4 import BeautifulSoup

html_text = requests.get(
    'https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation=')

soup = BeautifulSoup(html_text.text, 'lxml')
jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')

for job in jobs:
    company_name = job.find('h3', class_='joblist-comp-name')
    skills = job.find('div', class_='srp-skills')
    published_date = job.find('span', class_='sim-posted')

    if company_name and skills and published_date:
        company_name = company_name.text.strip()
        skills = ', '.join(skills.text.strip() for skill in skills.text.strip().split())
        published_date = published_date.span.text.strip()

        print(f'''   
    Company Name     : {company_name}
    Published Date   : {published_date}
    Required Skills  : {skills}
        ''')
        print("="*50)
```

```python
import requests
from bs4 import BeautifulSoup

html_text = requests.get(
    'https://wuzzuf.net/search/jobs?a=spbg&q=machine%20learning')

soup = BeautifulSoup(html_text.text, 'lxml')
jobs = soup.find_all('div', class_='css-pkv5jc')

for job in jobs:
    company_name = job.find('a', class_='css-17s97q8')
    keywords = job.find_all('a', class_='css-o171kl') + job.find_all('a', class_ = 'css-5x9pm1')
    for keyword in range(len(keywords)):
        keywords[keyword] = keywords[keyword].text
    published_date = job.find('div', class_='css-4c4ojb')
    Location = job.find('span', class_ = 'css-5wys0k')

    if company_name and keywords and published_date and Location:
        company_name = company_name.text.strip()
        Location = Location.text.strip()
        keywords = ', '.join(keywords)
        published_date = published_date.text.strip()

    print(f'''   
    Company Name     : {company_name}\b\b
    Location         : {Location}
    Published Date   : {published_date}
    keywords         : {keywords}

=======================================
    ''')
```

```python
import requests
from bs4 import BeautifulSoup
import csv
from time import sleep
from random import uniform


base_url = "https://wuzzuf.net/search/jobs"
query = "machine learning"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# ÿπÿØÿØ ÿßŸÑÿµŸÅÿ≠ÿßÿ™ ÿßŸÑŸÑŸä ŸáŸÜÿ¨ŸÖÿπ ŸÖŸÜŸáÿß ÿßŸÑŸàÿ∏ÿßÿ¶ŸÅ
number_of_pages = 5  

# ÿ™ÿ¨ŸáŸäÿ≤ ŸÖŸÑŸÅ CSV ŸÑÿ™ÿÆÿ≤ŸäŸÜ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
with open("wuzzuf_jobs.csv", mode="w", newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Job Title", "Company", "Location", "Posted", "Skills / Qualifications"])

    # ŸÜŸÑŸÅ ÿπŸÑŸâ ŸÉŸÑ ÿµŸÅÿ≠ÿ©
    for page in range(number_of_pages):
        params = {
            "q": query,
            "start": page * 15  # ŸÉŸÑ ÿµŸÅÿ≠ÿ© ŸÅŸä ÿßŸÑŸÖŸàŸÇÿπ ŸÅŸäŸáÿß 15 Ÿàÿ∏ŸäŸÅÿ© ÿ™ŸÇÿ±Ÿäÿ®Ÿãÿß
        }

        print(f"üîÑ Fetching page {page + 1}...")
        response = requests.get(base_url, headers=headers, params=params)
        soup = BeautifulSoup(response.content, "html.parser")

        jobs = soup.find_all("div", {"class": "css-1gatmva"})

        for job in jobs:
            title_tag = job.find("h2", {"class": "css-m604qf"})
            company_tag = job.find("a", {"class": "css-17s97q8"})
            location_tag = job.find("span", {"class": "css-5wys0k"})
            time_tag = job.find("div", {"class": "css-do6t5g"})
            skills_tag = job.find("div", {"class": "css-y4udm8"})

            title = title_tag.text.strip() if title_tag else "N/A"
            company = company_tag.text.strip() if company_tag else "N/A"
            location = location_tag.text.strip() if location_tag else "N/A"
            posted = time_tag.text.strip() if time_tag else "N/A"
            skills = skills_tag.text.strip().replace("\n", ", ") if skills_tag else "N/A"

            writer.writerow([title, company, location, posted, skills])

            print(f"‚úÖ {title} at {company} ({location})")

        # ÿ™ÿ£ÿÆŸäÿ± ÿπÿ¥Ÿàÿßÿ¶Ÿä ÿ®ŸäŸÜ ÿßŸÑÿµŸÅÿ≠ÿßÿ™ ÿπÿ¥ÿßŸÜ ŸÜÿ≠ÿßŸÉŸä ÿßŸÑÿ™ÿµŸÅÿ≠ ÿßŸÑÿ∑ÿ®ŸäÿπŸä 
            sleep(uniform(1.5, 3.0))

            print("\n Scraping complete! Data saved in 'wuzzuf_jobs.csv'")
```

```python
import tkinter as tk
from tkinter import ttk, scrolledtext, messagebox
import spacy
from spacy import displacy
import webbrowser
import tempfile
import os

# Load default model
nlp = spacy.load("en_core_web_sm")

def analyze_text():
    text = text_area.get("1.0", tk.END).strip()
    if not text:
        messagebox.showwarning("Warning", "Please enter some text.")
        return
    
    doc = nlp(text)
    
    # Clear old results
    for row in tree.get_children():
        tree.delete(row)
    
    # Insert entities into table
    if doc.ents:
        for ent in doc.ents:
            tree.insert("", "end", values=(ent.text, ent.label_, spacy.explain(ent.label_)))
    else:
        messagebox.showinfo("Result", "No entities found.")
    
    # Show displaCy visualization in browser
    html = displacy.render(doc, style="ent", jupyter=False)
    with tempfile.NamedTemporaryFile("w", delete=False, suffix=".html") as tmp:
        tmp.write(html)
        tmp_path = tmp.name
    webbrowser.open(f"file://{tmp_path}")

# --- Tkinter Window ---
root = tk.Tk()
root.title("spaCy Named Entity Recognition (NER)")

# Input area
tk.Label(root, text="Enter text for NER:").pack(anchor="w", padx=5, pady=5)
text_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=60, height=6)
text_area.insert(tk.END, "Apple was founded by Steve Jobs in California in 1976.")
text_area.pack(padx=5, pady=5, fill="both")

# Analyze button
analyze_btn = tk.Button(root, text="Extract Entities", command=analyze_text)
analyze_btn.pack(pady=5)

# Table for entities
columns = ("Entity", "Label", "Description")
tree = ttk.Treeview(root, columns=columns, show="headings", height=8)
for col in columns:
    tree.heading(col, text=col)
    tree.column(col, width=150)
tree.pack(padx=5, pady=5, fill="both", expand=True)

# Run Tkinter loop
root.mainloop()
```
